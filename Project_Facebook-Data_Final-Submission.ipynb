```python
# Defining function to find outliers from facebook dataset.
def detect_outliers(col,length):
    q1 = np.percentile(fbd[col],25)
    q3 = np.percentile(fbd[col],75)
    IQR = q3-q1
    
    # outlier step
    outlier_step = 1.5*IQR
  
    # Determining the list of indices of outliers for each feature column.
    outlier_list_col = fbd[(fbd[col] < q1 - outlier_step) | (fbd[col] > q3 + outlier_step)].index
  
    return outlier_list_col
```

```python
# Removing outliers in age column (users with age > 100 years, which seems not valid).
Age_Outliers_drop = detect_outliers('age',1.5)
fbd_Age_drop = fbd.drop(fbd.index[Age_Outliers_drop])
fbd = fbd_Age_drop
fbd.info()
```

```python
# Conversion of age column into groups for better visualization and analysis.
indices=[]
tIndices=[]
yIndices=[]
adoIndices=[]
adltIndices=[]
midageIndices=[]
sIndices=[]

for index, row in fbd.iterrows():
    if row['age']<13:
        indices.append(index)
        
    elif row['age']>=13 and row['age']<20:
        tIndices.append(index)
        
    elif row['age']>=20 and row['age']<25:
        yIndices.append(index)
        
    elif row['age']>=25 and row['age']<30:
        adoIndices.append(index)
        
    elif row['age']>=30 and row['age']<40:
        adltIndices.append(index)
        
    elif row['age']>=40 and row['age']<60:
        midageIndices.append(index)
        
    else:
        sIndices.append(index)
```

```python
# Categorically, age has been grouped in age_group column into child, teenagers, youngsters, adolescents, adults, mid-aged and senior-citizen based on some assumptions.
def age_group(row):
  if row['age'] < 13 :
     return 'Children'
  if row['age'] >= 13 and row['age'] < 20 :
     return 'Teenagers'
  if row['age'] >= 20 and row['age'] < 26 :
     return 'Youngsters'
  if row['age'] >= 26 and row['age'] < 31:
     return 'Adolescents'
  if row['age'] >= 31 and row['age'] < 41:
     return 'Adults'
  if row['age'] >= 41 and row['age'] < 61:
     return 'Mid-Aged'
  if row['age']  >= 61:
     return 'Senior-Citizen'
    
fbd['Age_Group'] = fbd.apply(lambda row: age_group(row),axis=1)
fbd.reset_index(drop=True, inplace=True)
fbd.info()

# infant (birth-12 months)
# toddler (1-4 years)
# child (5-11 years)
# preteen (12 years)
# teenager (13-19 years)
# youngster (20-25 years)
# adolescents (26-30 years)
# adult (31-40 years)
# mid-aged (41-60 years)
# senior citizen (60+ years)
```

```python
# Facebook data with age-gropus.
childernData = pd.DataFrame(fbd.ix[indices], columns = fbd.keys()).reset_index(drop=True)
teenagersData = pd.DataFrame(fbd.ix[tIndices], columns = fbd.keys()).reset_index(drop=True)
youngestersData = pd.DataFrame(fbd.ix[yIndices], columns = fbd.keys()).reset_index(drop=True)
adolescentsData = pd.DataFrame(fbd.ix[adoIndices], columns = fbd.keys()).reset_index(drop=True)
adltsData = pd.DataFrame(fbd.ix[adltIndices], columns = fbd.keys()).reset_index(drop=True)
midageData = pd.DataFrame(fbd.ix[midageIndices], columns = fbd.keys()).reset_index(drop=True)
sData = pd.DataFrame(fbd.ix[sIndices], columns = fbd.keys()).reset_index(drop=True)
```

```python
# Conversion of tenure column into Year groups for better visualization and analysis.
def Tenure_Year(row):
  if row['tenure'] <= 365 :
     return '1 year'
  if row['tenure'] > 365 and row['tenure'] <= 730 :
     return '2 year'
  if row['tenure'] > 730 and row['tenure'] <= 1095 :
     return '3 years'
  if row['tenure'] > 1095 and row['tenure'] <= 1460 :
     return '4 years'
  if row['tenure'] > 1460 and row['tenure'] <= 1825 :
     return '5 years'
  if row['tenure'] > 1825 and row['tenure'] <= 2190 :
     return '6 years'
  if row['tenure'] > 2190 and row['tenure'] <= 2555 :
     return '7 years'
  if row['tenure'] > 2555:
     return '8+ years'
  
fbd['Tenure_Year'] = fbd.apply(lambda row: Tenure_Year(row),axis=1)
fbd.reset_index(drop=True, inplace=True)
fbd.info()
```
